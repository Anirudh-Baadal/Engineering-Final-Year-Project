```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import Model,load_model 
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

# Loading the saved model

```python
model=load_model('../input/pn-model/model_2.h5')
```

# Custom image input

```python
image_path= '../input/chest-xray-pneumonia/chest_xray/test/PNEUMONIA/person119_bacteria_567.jpeg' 
```

# Prediction

```python
IMG_SIZE=150  
img = image.load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
plt.imshow(img,cmap="gray")
img=img_to_array(img)/255
img = np.expand_dims(img, axis=0)
result=model.predict(img) 
final_prediction=np.round(result[0][0])
if final_prediction==0:
    print("This person is free from pneumonia")
else:
    print("This person is infected by pneumonia")
```
